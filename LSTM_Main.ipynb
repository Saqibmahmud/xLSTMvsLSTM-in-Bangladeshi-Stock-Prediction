```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Load the dataset
df = pd.read_csv(r"D:\Research\Top10BDStocksData\AIL.csv")
df = df.dropna()
df['DATE'] = pd.to_datetime(df['DATE'])

# Shuffle the data (unsorting)
#df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Extracting close price
data = df[['CLOSEP*']].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Function to create sequences with different day differences
def create_sequences(data, seq_length, target_diff):
    X, y = [], []
    for i in range(len(data) - seq_length - target_diff):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length+target_diff])
    return np.array(X), np.array(y)

seq_length = 20
differences = [1, 3, 5, 10, 15]  # Days ahead prediction

results = {}

for diff in differences:
    print(f"\n### Checking accuracy for {diff}-day ahead prediction ###")
    
    X, y = create_sequences(data_scaled, seq_length, diff)

    # Split into training and testing sets
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Build the LSTM model
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(seq_length, 1)),
        Dropout(0.2),
        LSTM(50, return_sequences=False),
        Dropout(0.2),
        Dense(25, activation='sigmoid'),
        Dense(1)
    ])

    # Compile the model
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

    # Train the model
    es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=16, callbacks=[es], verbose=0)

    # Make predictions
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(predictions)  # Convert back to original scale
    y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
    print(df['DATE'].iloc[-len(y_test):],y_test_actual,predictions)
    # Compute error metrics
    mse = mean_squared_error(y_test_actual, predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test_actual, predictions)

    print(f"MSE: {mse:.4f}, RMSE: {rmse:.4f}, MAE: {mae:.4f}")

    results[diff] = {"MSE": mse, "RMSE": rmse, "MAE": mae}

    # Plot results
    plt.figure(figsize=(10, 5))
    plt.plot(df['DATE'].iloc[-len(y_test):], y_test_actual, label='Actual Prices', color='blue')
    plt.plot(df['DATE'].iloc[-len(y_test):], predictions, label='Predicted Prices', color='red')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.title(f"{diff}-Day Prediction")
    plt.legend()
    plt.show()

# Print results summary
print("\nFinal Results Summary:")
for diff, metrics in results.items():
    print(f"{diff}-Day Ahead Prediction -> MSE: {metrics['MSE']:.4f}, RMSE: {metrics['RMSE']:.4f}, MAE: {metrics['MAE']:.4f}")
```
